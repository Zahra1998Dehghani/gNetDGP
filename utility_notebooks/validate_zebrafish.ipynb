```python
import pandas as pd
```

```python
all_diseases = pd.read_csv("all_diseases.tsv", sep="\t", header=None)
all_diseases
```

```python
gene_diseases = pd.read_csv("genes_diseases.tsv", sep="\t", header=None)
gene_diseases
```

```python
gene2diseases_zebrafish = pd.read_csv("gene2DiseaseViaOrthology_2024.04.03.tsv", sep="\t")
gene2diseases_zebrafish
```

```python
zebrafish_human_gene_ids = list(set(gene2diseases_zebrafish["Human_Ortholog_Entrez_Gene_Id"].tolist()))
len(zebrafish_human_gene_ids)
```

```python
human_genes = list(set(gene_diseases.loc[:, 0].tolist()))
len(human_genes)
```

```python
gene_ids_not_present = list(set(zebrafish_human_gene_ids).difference(set(human_genes)))
len(gene_ids_not_present)
```

```python
all_disease_names = all_diseases.loc[:, 1].tolist()
all_disease_OMIM = list(set([int(item.split(":")[1]) for item in all_disease_names]))
len(all_disease_OMIM), all_disease_names[:5], all_disease_OMIM[:5]
```

```python
import numpy as np
zebrafish_OMIM_ID = list(set(gene2diseases_zebrafish["OMIM_ID"].tolist()))
zebrafish_OMIM_ID = [int(item) for item in zebrafish_OMIM_ID if item > 0]
len(zebrafish_OMIM_ID), zebrafish_OMIM_ID[:5]
```

```python
disease_ids_not_present = list(set(zebrafish_OMIM_ID).difference(set(all_disease_OMIM)))
len(disease_ids_not_present), disease_ids_not_present[:5]
```

```python
gene2diseases_zebrafish_filtered = gene2diseases_zebrafish[~gene2diseases_zebrafish["Human_Ortholog_Entrez_Gene_Id"].isin(gene_ids_not_present)]
gene2diseases_zebrafish_filtered
```

```python
gene2diseases_zebrafish_filtered["OMIM_ID"] = gene2diseases_zebrafish_filtered["OMIM_ID"].fillna(-1)
gene2diseases_zebrafish_filtered["OMIM_ID"] = gene2diseases_zebrafish_filtered["OMIM_ID"].astype(int)
gene2diseases_zebrafish_filtered
```

```python
gene2diseases_zebrafish_genes_OMIMs = gene2diseases_zebrafish_filtered[["Human_Ortholog_Entrez_Gene_Id", "OMIM_ID"]]
gene2diseases_zebrafish_genes_OMIMs
```

```python
gene2diseases_zebrafish_genes_OMIMs["OMIM_ID"] =  'OMIM:' + gene2diseases_zebrafish_genes_OMIMs['OMIM_ID'].astype(str)
gene2diseases_zebrafish_genes_OMIMs
```

```python
gene2diseases_zebrafish_genes_OMIMs.columns = range(gene2diseases_zebrafish_genes_OMIMs.shape[1])
gene2diseases_zebrafish_genes_OMIMs
#gene2diseases_zebrafish_genes_OMIMs_nh.to_csv("gene2diseases_zebrafish_genes_OMIMs.tsv", index=None)
```

```python
unique_gene2diseases_zebrafish_genes_OMIMs = gene2diseases_zebrafish_genes_OMIMs.drop_duplicates()
unique_gene2diseases_zebrafish_genes_OMIMs
```

```python
unique_gene_diseases = gene_diseases.drop_duplicates()
unique_gene_diseases
```

```python
concat_pd = pd.concat([unique_gene2diseases_zebrafish_genes_OMIMs, gene_diseases], ignore_index=True, axis=0)
concat_pd
```

```python
concat_pd.to_csv("human_zebrafish_genes_diseases.tsv", index=None, sep="\t")
```

```python
unique_concat_pd = concat_pd.drop_duplicates()
unique_concat_pd
```
