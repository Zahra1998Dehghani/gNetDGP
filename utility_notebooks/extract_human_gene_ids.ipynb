```python
import pandas as pd

df_genes = pd.read_csv("all_human_genes.txt", sep="\t")
df_genes
```

```python
df_human_genes = df_genes[df_genes["Org_name"] == "Homo sapiens"]
df_human_genes
```

```python
df_human_genes_data = df_human_genes[["Symbol", "GeneID"]]
df_human_genes_data
```

```python
#df_human_genes_data.to_csv("df_human_genes_data_NCBI.tsv", sep="\t", index=None)
```

```python
human_gene_ids = pd.read_csv("uniprotKB_human_geneIDS.tsv", sep="\t")
human_gene_ids
```

```python
human_genes = human_gene_ids[["Gene Names", "GeneID"]]
human_genes = human_genes.dropna()
human_genes
```

```python
for i, row in human_genes.iterrows():
    #print(row[0], row[1])
    try:
        row[0] = row[0].split(" ")[0]
        row[1] = row[1].split(" ")[1]
    except:
        continue
human_genes
```

```python
human_genes.columns = ["Symbol", "GeneID"]
human_genes
```

```python
geneids = [int(item.replace(";", "")) for item in human_genes["GeneID"].tolist()]
human_genes["GeneID"] = geneids
human_genes
```

```python
combined_human_genes = pd.concat([df_human_genes_data, human_genes], axis=0)
combined_human_genes
```

```python
combined_human_genes_unique = combined_human_genes.drop_duplicates()
combined_human_genes_unique
```
