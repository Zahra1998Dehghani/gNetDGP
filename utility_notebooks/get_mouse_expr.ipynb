```python
gene_names = mouse_gene_expr_only_ensemble.iloc[:, 1].tolist()
unique_mouse_genes = list(set(gene_names))
len(unique_mouse_genes)
```

```python
from Bio import Entrez

def get_entrez_gene_id(gene_name):
    Entrez.email = 'anup.rulez@gmail.com'  # Provide your email for NCBI API usage
    handle = Entrez.esearch(db='gene', term=gene_name)
    record = Entrez.read(handle)
    handle.close()
    if record['IdList']:
        return record['IdList'][0]  # Return the first matching Entrez Gene ID
    else:
        return None  # Gene not found


mouse_gene_ids = list()
mouse_gene_names = list()

for gene_name in unique_mouse_genes:
    entrez_gene_id = get_entrez_gene_id(gene_name)
    if entrez_gene_id:
        mouse_gene_ids.append(entrez_gene_id)
        mouse_gene_names.append(gene_name)
        
mouse_gene_info = pd.DataFrame(zip(mouse_gene_ids, mouse_gene_names), columns=["GeneID", "GeneName"])
mouse_gene_info
```
